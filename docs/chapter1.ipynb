## Battery Management with ROS 2 Humble

In this tutorial, we aim to demonstrate how to develop from scratch a testing functionality for a battery management used in the navigation of an indoor autonomous robot operating in an industrial-like environment, the **J16 MOTO robot**.
The objective is to provide a practical guide to developing a simple application in ROS 2, illustrating the main communication mechanisms and node management concepts required to interface with the robot’s sensors and verify their correct operation within a real or simulated robotic system.

To achieve this, several fundamental ROS 2 concepts will be introduced and applied to the amrj16_ultimac platform, which provide the foundation for developing modular and scalable robotic applications.

In particular, the tutorial will cover the following topics:

1. Basic ROS 2 concepts: packages, launch files, nodes, and clients.

2. Topic Pub/Sub: the communication mechanism between nodes based on message publishing and subscription
3. Custom interfaces: how to define custom message types to structure and exchange data between nodes.

4. Service Client/Server: how to request the robot to perform an operation and receive a synchronous or asynchronous response.

5. Action Client/Server: how to manage long-running tasks while receiving progress updates.

6. Multithreading: techniques that allow multiple operations to be executed concurrently within a node.

7. Debug tools: the use of RViz and logging systems to monitor and analyze the robot’s behavior.

By the end of this tutorial, the reader will have acquired the fundamental concepts needed to develop and test several functionalities based on ROS 2, applicable to the development of the J16 MOTO robot in industrial environments.
